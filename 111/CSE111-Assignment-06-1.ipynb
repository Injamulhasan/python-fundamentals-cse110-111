```python
#task 01




class Student:
    ID = 0
    def __init__(self, name, dept, age, cgpa):
        self.name = name
        self.dept = dept
        self.age = age
        self.cgpa = cgpa
        Student.ID += 1
        
    def get_details(self):
        print("ID:", Student.ID)
        print("Name:", self.name)
        print("Department:", self.dept)
        print("Age:", self.age)
        print("CGPA:", self.cgpa)
    
    @classmethod
    def from_String(cls, info): 
        name, dept, age, cgpa = info.split("-")
        obj = cls(name, dept, age, cgpa)
        return obj
    
s1 = Student("Samin", "CSE", 21, 3.91)
s1.get_details()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.get_details()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.get_details()
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.get_details() 

print(" ")
print("The difference between a class variable and an instance variable is that class variables are shared amongst all objects of the class while the instance variables are different for each object of a class.")
print(" ")
print("The difference between an instance method and a class method is that instance methods can access the attributes and the other methods through the 'self' parameter while the class method takes a different parameter 'cls' that points to the class instead of the object.")




```

```python
#task 02



class Assassin:    
    totalAssasinCount = 0    
    def __init__(self, name, successRate):        
        self.name = name
        self.successRate = successRate
        Assassin.totalAssasinCount += 1    
    @classmethod    
    def failureRate(cls, name, failureCount):
        return cls(name, 100 - failureCount)
        Assassin.totalAssasinCount += 1

    @classmethod
    def failurePercentage(cls, name, failurePercentage):
        return cls(name, 100 - failurePercentage)
        Assassin.totalAssasinCount += 1
    
    
    def printDetails(self):
        print("Name: {}".format(self.name))
        print("Success rate: {}%".format(self.successRate))
        print("Total number of Assassin: {}".format(self.totalAssasinCount))
        
john_wick = Assassin('John Wick', 100)
john_wick.printDetails()      
print("================================")

nagisa = Assassin.failureRate('Nagisa', 20)
nagisa.printDetails()
print("================================")
akabane = Assassin.failurePercentage('Akabane', 10)
akabane.printDetails()



```

```python
#task 03




class Passenger:
    count = 0
    
    def __init__(self, name, bus_fare = 450):
        self.name = name
        self.bus_fare = bus_fare
        Passenger.count += 1
            
    def printDetail(self):
        print("Name:", self.name)
        print("Bus Fare:", self.bus_fare, "taka")
    
    def set_bag_weight(self, bag_weight):
        if bag_weight <= 20:
            self.bus_fare += 0
        elif 21 <= bag_weight <= 50:
            self.bus_fare += 50
        elif bag_weight > 50:
            self.bus_fare += 100
        return self.bus_fare
    
    
print("Total Passenger:", Passenger.count)
p1 = Passenger("Jack")
p1.set_bag_weight(90)
p2 = Passenger("Carol")
p2.set_bag_weight(10)
p3 = Passenger("Mike")
p3.set_bag_weight(25)
print("=========================")
p1.printDetail()
print("=========================")
p2.printDetail()
print("=========================")
p3.printDetail()
print("=========================")
print("Total Passenger:", Passenger.count)




```

```python
#5




from datetime import date


class Employee:    
    def __init__(self, name, workPeriod):
        self.name = name
        self.workingPeriod = workPeriod
    
    @classmethod    
    def employeeByJoiningYear(cls, name, workPeriod):
        return cls(name, date.today().year - workPeriod)

    @staticmethod
    def experienceCheck(workPeriod, gender):
        expStatus = ''
        if workPeriod < 3:
            expStatus = " is not experienced"
        else:
            expStatus = " is experienced"
        
        if gender.upper() == "MALE":
            return "He" + expStatus
        else:
            return "She" + expStatus
   
        
employee1 = Employee('Dororo', 3)    
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, 'male'))
print(Employee.experienceCheck(3, 'female'))



```

```python
#task 06





class Laptop:
    laptopcount = 0
    
    def __init__(self, name, count):
        self.name = name
        self.count = count
        Laptop.laptopcount += self.count
        
    @classmethod
    def advantage(cls):
        print("Laptop are portable")
    
    @classmethod
    def resetcount(cls):
        cls.laptopcount = 0
        
    
    
lenovo = Laptop("Lenovo", 5);
dell = Laptop("Dell", 7);
print(lenovo.name, lenovo.count)
print(dell.name,dell.count)
print("Total number of Laptop", Laptop.laptopcount)
Laptop.advantage()
Laptop.resetcount()
print("Total number of Laptop", Laptop.laptopcount)



```

```python
#task 07





class Cat:
    Number_of_cats = 0
    
    def __init__(self, color, action):
        self.color = color
        self.action = action
        Cat.Number_of_cats += 1
        
    @classmethod 
    def no_parameter(cls, up_color = "White", up_action = "sitting"):
        return cls(up_color, up_action)
    
    @classmethod
    def first_parameter(cls, up_color, up_action = "sitting"):
        return cls(up_color, up_action)
        
    @classmethod
    def second_parameter(cls, up_action, up_color = "Grey"):
        return cls(up_color, up_action)
    
    def printCat(self):
        print(self.color, "cat is", self.action)
        
    def changeColor(self, color):
        self.color = color
        


print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)



```

```python
#task 08





import math
class Cylinder:
    radius = 5
    height = 18
    
    def __init__(self, radius, height):
        print("Default radius =", Cylinder.radius, "and height =", Cylinder.height)
        Cylinder.radius = radius
        Cylinder.height = height
        print("Updated radius =", Cylinder.radius, "and height =", Cylinder.height)
        
    @classmethod
    def swap(cls, up_height, up_radius):
        return cls(up_radius, up_height)
    
    @classmethod
    def changeFormat(cls, val):
        up_radius, up_height = val.split("-")
        up_radius = float(up_radius)
        up_height = float(up_height)
        return cls(up_radius, up_height)
        
    @staticmethod
    def area(radius, height):
        cal_area = (2 * math.pi * (radius**2)) + (2 * math.pi * radius * height)
        print("Area:", cal_area)
        
    @staticmethod
    def volume(radius, height):
        cal_volume = math.pi * (radius**2) * height
        print("Volume:", cal_volume)
    

c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)




```
