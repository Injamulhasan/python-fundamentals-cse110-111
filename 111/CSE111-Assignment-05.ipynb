```python
#task 01




class Marks:
    def __init__(self, marks):
        self.mark = marks
    def mark(self):
        return self.mark
    def __add__(self, other):
        result = self.mark + other.mark
        return Marks(result)
    

Q1 = Marks(int(input("Quiz 1 (out of 10): ")))
Q2 = Marks(int(input("Quiz 2 (out of 10): ")))
Lab = Marks(int(input("Lab (out of 30): ")))
Mid = Marks(int(input("Mid (out of 20): ")))
Final = Marks(int(input("Final (out of 30): ")))
total = Q1 + Q2 + Lab + Mid + Final
print("Total marks: {}".format(total.mark))
```

```python
#task 02




class Teacher:
    def __init__(self, name, dept):
        self.__name = name
        self.__dept = dept
        self.course = []
        
    def addCourse(self, courses):
        self.course.append(courses.course)

    def printDetail(self):
        print("====================================")
        print("Name:", self.__name)
        print("Department:", self.__dept)
        print("List of courses")
        print("====================================")
        for i in self.course:
            print(i)
        print("====================================")
    
class Course:
    def __init__(self, course_name):
        self.course = course_name
        

t1 = Teacher("Saad Abdullah", "CSE")
t2 = Teacher("Mumit Khan", "CSE")
t3 = Teacher("Sadia Kazi", "CSE")
c1 = Course("CSE 110 Programming Language I")
c2 = Course("CSE 111 Programming Language-II")
c3 = Course("CSE 220 Data Structures")
c4 = Course("CSE 221 Algorithms")
c5 = Course("CCSE 230 Discrete Mathematics")
c6 = Course("CSE 310 Object Oriented Programming")
c7 = Course("CSE 320 Data Communications")
c8 = Course("CSE 340 Computer Architecture")
t1.addCourse(c1)
t1.addCourse(c2)
t2.addCourse(c3)
t2.addCourse(c4)
t2.addCourse(c5)
t3.addCourse(c6)
t3.addCourse(c7)
t3.addCourse(c8)
t1.printDetail()
t2.printDetail()
t3.printDetail()
```

```python
#task 04



class Color:
    def __init__(self, color):
        self.clr = color
        
    def __add__(self, other):
        if (self.clr.lower() == "red" and other.clr.lower() == "yellow") or (self.clr.lower() == "yellow" and other.clr.lower() == "red"):
            return Color("Orange")
        elif (self.clr.lower() == "red" and other.clr.lower() == "blue") or (self.clr.lower() == "blue" and other.clr.lower() == "red"):
            return Color("Violet")
        elif (self.clr.lower() == "yellow" and other.clr.lower() == "blue") or (self.clr.lower() == "blue" and other.clr.lower() == "yellow"):
            return Color("Green")
        else:
            pass
        
C1 = Color(input("First Color: ").lower())
C2 = Color(input("Second Color: ").lower())
C3 = C1 + C2
print("Color formed:", C3.clr)
```

```python
#task 05



import math
class Circle:
    def __init__(self, radius):
        self.__radius = radius
        
    def __add__(self, other):
        result = self.__radius + other.__radius
        return Circle(result)
        
    def setRadius(self):
        self.__radius = result
            
    def getRadius(self):
        return self.__radius
    
    def area(self):
        circle_area = math.pi * (self.__radius ** 2)
        return circle_area
    


c1 = Circle(4)
print("First circle radius:", c1.getRadius())
print("First circle area:", c1.area())
c2 = Circle(5)
print("Second circle radius:", c2.getRadius())
print("Second circle area:", c2.area())
c3 = c1 + c2
print("Third circle radius:", c3.getRadius())
print("Third circle area:", c3.area())
```

```python
#task 06




class Triangle:
    def __init__(self, base, height):
        self.__base = base
        self.__height = height
        
    def __sub__(self, other):
        new_base = self.__base - other.__base
        new_height = self.__height - other.__height
        return Triangle(new_base, new_height) 
        
    def setBase(self):
        self.__base = new_base
            
    def getBase(self):
        return self.__base
    
    def setHeight(self):
        self.__height = new_height
            
    def getHeight(self):
        return self.__height
    
    def area(self):
        triangle_area = 0.5 * self.__base * self.__height
        return float(triangle_area)

t1 = Triangle(10, 5)
print("First Triangle Base:", t1.getBase())
print("First Triangle Height:", t1.getHeight())
print("First Triangle area:", t1.area())
t2 = Triangle(5, 3)
print("Second Triangle Base:", t2.getBase())
print("Second Triangle Height:", t2.getHeight())
print("Second Triangle area:", t2.area())
t3 = t1 - t2
print("Third Triangle Base:", t3.getBase())
print("Third Triangle Height:", t3.getHeight())
print("Third Triangle area:", t3.area())
```

```python
#task 07




class Dolls:
    def __init__(self, doll, price):
        self.doll = doll
        self.price = price
    
    def detail(self):
        return "Doll: " + self.doll + "\n" + "Total Price: " + str(self.price) + " taka"
    
    def __add__(self, other):
        new_doll = self.doll + " " + other.doll
        new_price = self.price + other.price
        return Dolls(new_doll, new_price)
    
    def __gt__(self, other):
        if self.price > other.price:
            return True
        else:
            return False
    
    
obj_1 = Dolls("Tweety", 2500)
print(obj_1.detail())
if obj_1 > obj_1:
    print("Congratulations! You get the Tweety as a gift!")
else:
    print("Thank you!")
print("=========================")
obj_2 = Dolls("Daffy Duck", 1800)
print(obj_2.detail())
if obj_2 > obj_1:
    print("Congratulations! You get the Tweety as a gift!")
else:
    print("Thank you!")
print("=========================")
obj_3 = Dolls("Bugs Bunny", 3000)
print(obj_3.detail())
if obj_3 > obj_1:
    print("Congratulations! You get the Tweety as a gift!")
else:
    print("Thank you!")
print("=========================")
obj_4 = Dolls("Porky Pig", 1500)
print(obj_4.detail())
if obj_4 > obj_1:
    print("Congratulations! You get the Tweety as a gift!")
else:
    print("Thank you!")
print("=========================")
obj_5 = obj_2 + obj_3
print(obj_5.detail())
if obj_5 > obj_1:
    print("Congratulations! You get the Tweety as a gift!")
else:
    print("Thank you!")
```

```python
#task 08




class Coordinates:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2
        
    def __sub__(self, other):
        cal1 = self.num1 - other.num1
        cal2 = self.num2 - other.num2
        output = Coordinates(cal1, cal2)
        return output
    
    def __mul__(self, other):
        cal1 = self.num1 * other.num1
        cal2 = self.num2 * other.num2
        output = Coordinates(cal1, cal2)
        return output
    
    def __eq__(self, other):
        if (self.num1 == other.num1) and (self.num2 == other.num2):
            return "The calculated coordinates are the same."
        else:
            return "The calculated coordinates are NOT the same."
    
    def detail(self):
        return self.num1, self.num2


p1 = Coordinates(int(input()),int(input()))
p2 = Coordinates(int(input()),int(input()))
p4 = p1 - p2
print(p4.detail())
p5 = p1 * p2
print(p5.detail())
point_check = (p4 == p5)
print(point_check)
```
