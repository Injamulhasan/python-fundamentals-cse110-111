```python
#task 1










class Calculator:
    def __init__(self, num1, operator, num2):
        self.num1 = num1
        self.operator = operator 
        self.num2 = num2
        print("Let's Calculate!")
        print("Value 1:", self.num1)
        print("Operator:", self.operator)
        print("Value 2:", self.num2)

    def add(self):
        result = self.num1 + self.num2
        print("Result:", result)

    def subtract(self):
        result = self.num1 - self.num2
        print("Result:", result)

    def multiply(self):
        result = self.num1 * self.num2
        print("Result:", result)

    def divide(self):
        result = self.num1 / self.num2
        print("Result:", result)

input_num1 = int(input("Enter first number: "))
input_operator = input("Enter operator: ")
input_num2 = int(input("Enter second number: "))
c1 = Calculator(input_num1, input_operator, input_num2)

if input_operator == "+":
    c1.add()
elif input_operator == "-":
    c1.subtract()
elif input_operator == "*":
    c1.multiply()
elif input_operator == "/":
    c1.divide()
```

```python
#task 2







class Customer:
    def __init__(self, name):
        self.name = name
        
    def greet(self, name = None):
        if name != None:
            print("Hello "+ name + "!")
        else:
            print("Hello!")
        
    def purchase(self, *items):
        print(self.name, "you purchased", len(items), "item(s):")
        for x in items:
            print(x)
            
customer_1 = Customer("Sam")
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice")
print("-----------------------------")
customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice")
```

```python
#task 3








class Panda:
    def __init__(self, name, gender, age):
        self.name = name
        self.gender = gender
        self.age = age
    def sleep(self, time = None):
        if time != None:
            if 3 <= time <= 5:
                return f"{self.name} sleeps {time} hours daily and should have Mixed Veggies"
            elif 6 <= time <= 8:
                return f"{self.name} sleeps {time} hours daily and should have Eggplant & Tofu"
                return "Eggplant & Tofu"
            elif 9 <= time <= 11:
                return f"{self.name} sleeps {time} hours daily and should have Broccoli Chicken"
                return "Broccoli Chicken"
        else:
            return f"{self.name}'s duration is unknown thus should have only bamboo leaves"

panda1 = Panda("Kunfu","Male", 5)
panda2=Panda("Pan Pan","Female",3)
panda3=Panda("Ming Ming","Female",8)
print("{} is a {} Panda Bear who is {} years old".format(panda1.name,panda1.gender,panda1.age))
print("{} is a {} Panda Bear who is {} years old".format(panda2.name,panda2.gender,panda2.age))
print("{} is a {} Panda Bear who is {} years old".format(panda3.name,panda3.gender,panda3.age))
print(panda2.sleep(10))
print(panda1.sleep(4))
print(panda3.sleep())
```

```python
#task 4






class Cat:
    def __init__(self, *info):
        if len(info) == 2:
            self.color = info[0]
            self.action = info[1]
        elif len(info) == 1:
            self.color = info[0]
            self.action = "sitting"
        elif len(info) == 0:
            self.color = "White"
            self.action = "sitting"
            
    def printCat(self):
        print(self.color, "cat is", self.action)
        
    def changeColor(self, color):
        self.color = color
        
c1 = Cat()
c2 = Cat("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
```

```python
#task 5








class Vehicle:
    def __init__(self):
        self.x_coordinate = 0
        self.y_coordinate = 0
    def moveUp(self):
        self.x_coordinate = 0
        self.y_coordinate = 1
    def moveLeft(self):
        self.x_coordinate = -1
        self.y_coordinate = 1
    def moveDown(self):
        self.x_coordinate = -1
        self.y_coordinate = 0
    def moveRight(self):
        self.x_coordinate = 0
        self.y_coordinate = 0
    def print_position(self):
        output = self.x_coordinate, self.y_coordinate
        print(output)
    
     
car = Vehicle()
car.print_position()
car.moveUp()
car.print_position()
car.moveLeft()
car.print_position()
car.moveDown()
car.print_position()
car.moveRight()    
```

```python
#task 6








class Programmer:
    def __init__(self, name, language, experience):
        self.name = name
        self.language = language
        self.experience = experience
        print("Horray! A new programmer is born")
    def addExp(self, updated_experience = None):
        if updated_experience != None:
            self.experience += updated_experience
            print("Updating experience of", self.name)
        else:
            pass
    def printDetails(self):
        print("Name:", self.name)
        print("Language:", self.language)
        print("Experience:", self.experience, "years")
    

    
p1 = Programmer("Ethen Hunt", "Java", 10)
p1.printDetails()
print('--------------------------')
p2 = Programmer("James Bond", "C++", 7)
p2.printDetails()
print('--------------------------')
p3 = Programmer("Jon Snow", "Python", 4)
p3.printDetails()
p3.addExp(5)
p3.printDetails() 
```

```python
#task 7








class Student:
    def __init__(self, name, Id, department = "CSE"):
        self.name = name
        self.id = Id
        self.department = department
    def printDetails(self):
        print("Name:", self.name)
        print("ID:", self.id)
        print("Department:", self.department)
    def dailyEffort(self, hour):
        print("Daily Effort:", hour, "hour(s)")
        if hour <= 2:
            print("Suggestion: Should give more effort!")
        elif hour <= 4:
            print("Suggestion: Keep up the good work!")
        else:
            print("Suggestion: Excellent! Now motivate others.")
    
    
    
harry = Student('Harry Potter', 123)
harry.printDetails()
harry.dailyEffort(3)
print('========================')
john = Student("John Wick", 456, "BBA")
john.printDetails()
john.dailyEffort(2)
print('========================')
naruto = Student("Naruto Uzumaki", 777, "Ninja")
naruto.printDetails()
naruto.dailyEffort(6)
```

```python
#task 8







class Patient:
    def __init__(self, name, age):
        self.name = name 
        self.age = age
        self.symptom = ""
    def add_Symptom(self, *symptoms):
        for i in symptoms:
            if len(symptoms) == 1:
                self.symptom += i
            else:
                self.symptom = self.symptom + i + ", "
    def printPatientDetail(self):
        print("Name:", self.name)
        print("Age:", self.age)
        print("Symptoms:", self.symptom)
        
p1 = Patient("Thomas", 23)
p1.add_Symptom("Headache")
p2 = Patient("Carol", 20)
p2.add_Symptom("Vomiting", "Coughing")
p3 = Patient("Mike", 25)
p3.add_Symptom("Fever", "Headache", "Coughing")
print("=========================")
p1.printPatientDetail()
print("=========================")
p2.printPatientDetail()
print("=========================")
p3.printPatientDetail()
print("=========================")
```

```python
#task 9






class Avengers:
    def __init__(self, name, partner):
        self.name = name
        self.partner = partner
        self.super_power = ""
    def super_powers(self, *super_power):
        for i in super_power:
            if len(super_power) == 1:
                self.super_power += i
            else:
                self.super_power = self.super_power + i + ", "
    def printAvengersDetail(self):
        print("Name:", self.name)
        print("Age:", self.partner)
        print("Symptoms:", self.super_power)    
    
    
    
    
a1 = Avengers('Captain America', 'Bucky Barnes')
a1.super_powers('Stamina', 'Slowed ageing')
a2 = Avengers('Doctor Strange', 'Ancient One')
a2.super_powers('Mastery of magic')
a3 = Avengers('Iron Man', 'War Machine')
a3.super_powers('Genius level intellect', 'Scientist ')
print("=========================")
a1.printAvengersDetail()
print("=========================")
a2.printAvengersDetail()
print("=========================")
a3.printAvengersDetail()
print("=========================")
```

```python
#task 10








class Shinobi:
    def __init__(self, name, rank, mission = 0, salary = 0):
        self.name = name 
        self.rank = rank
        self.mission = mission
        self.salary = salary
    def changeRank(self, new_rank):
        self.rank = new_rank    
    def calSalary(self, new_mission):
        self.mission = new_mission
        if self.rank == "Genin":
            self.salary = new_mission * 50
        elif self.rank == "Chunin":
            self.salary = new_mission * 100
        else:
            self.salary = new_mission * 500
    def printInfo(self):
        print("Name:", self.name)
        print("Rank:", self.rank)
        print("Number of mission:", self.mission)
        print("Salary:", self.salary)
    
naruto = Shinobi("Naruto", "Genin")
naruto.calSalary(5)
naruto.printInfo()
print('====================')
shikamaru = Shinobi('Shikamaru', "Genin")
shikamaru.printInfo()
shikamaru.changeRank("Chunin")
shikamaru.calSalary(10)
shikamaru.printInfo()
print('====================')
neiji = Shinobi("Neiji", "Jonin")
neiji.calSalary(5)
neiji.printInfo()
```
