```python
#task 01




class RealNumber:
    def __init__(self, r=0):
        self.__realValue = r
    def getRealValue(self):
        return self.__realValue
    def setRealValue(self, r):
        self.__realValue = r
    def __str__(self):
        return 'RealPart: '+str(self.getRealValue())
    
    
class ComplexNumber(RealNumber):
    def __init__(self, r = 1.0, i = 1.0):
        super().__init__(r)
        self.imaginaryValue = float(i)
                   
    def __str__(self):
        return "RealPart: " + str(float(super().getRealValue())) + "\n" + "ImaginaryPart: " + str(self.imaginaryValue)

    
cn1 = ComplexNumber()
print(cn1)
print('---------')
cn2 = ComplexNumber(5,7)
print(cn2)
```

```python
#task 02



class RealNumber:
    def __init__(self, number=0):
        self.number = number
    def __add__(self, anotherRealNumber):
        return self.number + anotherRealNumber.number
    def __sub__(self, anotherRealNumber):
        return self.number - anotherRealNumber.number
    def __str__(self):
        return str(self.number)
    
    
class ComplexNumber(RealNumber):
    def __init__(self, r, i):
        self.imaginaryValue = i
        if isinstance(r, RealNumber):
            self.realValue = r.number
        else:
            self.realValue = r
            
    def __add__(self, other):
        sum1 = self.realValue + other.realValue
        sum2 = self.imaginaryValue + other.imaginaryValue
        return ComplexNumber(sum1, sum2)
    
    def __sub__(self, other):
        sum1 = self.realValue - other.realValue
        sum2 = self.imaginaryValue - other.imaginaryValue
        return ComplexNumber(sum1, sum2)
    
    def __str__(self):
        if self.imaginaryValue >= 0:
            operator = "+"
        else:
            operator = "-"
        return str(self.realValue) + " " + operator + " "+ str(abs(self.imaginaryValue)) + "i"
    

r1 = RealNumber(3)
r2 = RealNumber(5)
print(r1+r2)
cn1 = ComplexNumber(2, 1)
print(cn1)
cn2 = ComplexNumber(r1, 5)
print(cn2)
cn3 = cn1 + cn2
print(cn3)
cn4 = cn1 - cn2
print(cn4)
```

```python
#task 03




class Account:
    def __init__(self, balance):
        self._balance = balance
    def getBalance(self):
        return self._balance
    
    
class CheckingAccount(Account):
    numberOfAccount = 0
    
    def __init__(self, balance = 0.0):
        super().__init__(balance)
        CheckingAccount.numberOfAccount += 1
        
    def __str__(self):
        if self._balance == 0.0:
            return "Account Balance: " + str(self._balance)
        else:
            return "Account Balance: " + "{:.2f}".format(self._balance)
        

print('Number of Checking Accounts: ', CheckingAccount.numberOfAccount)
print(CheckingAccount())
print(CheckingAccount(100.00))
print(CheckingAccount(200.00))
print('Number of Checking Accounts: ', CheckingAccount.numberOfAccount)
```

```python
#task 05





class Exam:
    def __init__(self,marks):
        self.marks = marks
        self.time = 60
    def examSyllabus(self):
        return "Maths , English"
    def examParts(self):
        return "Part 1 - Maths\nPart 2 - English\n"
    

class ScienceExam(Exam):
    parts = 2
    def __init__(self, marks, time, part1, part2, part3 = None):
        super().__init__(marks)
        self.time = time
        self.part1 = part1
        self.part2 = part2
        self.part3 = part3
        if self.part3 == None:
            ScienceExam.parts = ScienceExam.parts + 2
        else:
            ScienceExam.parts = ScienceExam.parts + 3
        
            
    def examSyllabus(self):
        print(super().examSyllabus(), end= " , ")
        if self.part3 == None:
            return self.part1 + " , " + self.part2
        else:
            return self.part1 + " , " + self.part2 + " , " + self.part3
    
    def examParts(self):
        print(super().examParts(), end = "")
        if self.part3 == None:
            return "Part 3 - " + self.part1 + "\nPart 4 - " + self.part2
        else:
            return "Part 3 - " + self.part1 + "\nPart 4 - " + self.part2 + "\nPart 5 - " + self.part3
    
    def __str__(self):
        return "Marks: 100 " + "Time: " + str(self.time) + " minutes" + " Number of Parts: " + str(ScienceExam.parts)
              

engineering = ScienceExam(100,90,"Physics","HigherMaths")
print(engineering)
print('----------------------------------')
print(engineering.examSyllabus())
print(engineering.examParts())
print('==================================')
architecture = ScienceExam(100,120,"Physics","HigherMaths","Drawing")
print(architecture)
print('----------------------------------')
print(architecture.examSyllabus())
print(architecture.examParts())
```

```python
#task 06





class Shape3D:
    pi = 3.14159
    def __init__(self, name = 'Default', radius = 0):
        self._area = 0
        self._name = name
        self._height = 'No need'
        self._radius = radius
    def calc_surface_area(self):
        return 2 * Shape3D.pi * self._radius
    def __str__(self):
        return "Radius: "+str(self._radius)
    
    

class Sphere(Shape3D):  
    def __init__(self, name = 'Default', radius = 0):
        super().__init__(name, radius)
        print("Shape name: " + self._name + ", Area Formula: 4 * pi * r * r")
        
    def calc_surface_area(self):
        return 4 * Shape3D.pi * self._radius * self._radius
        
    def __str__(self):
        print(super().__str__(), end = ", ")
        return "Height: " + self._height + "\nArea: " + str(self.calc_surface_area())
                
        
class Cylinder(Shape3D):  
    def __init__(self, name = 'Default', radius = 0, height = 0):
        super().__init__(name, radius)
        self._height = height
        print("Shape name: " + self._name + ", Area Formula: 2 * pi * r * (r + h)")
        
    def calc_surface_area(self):
        return 2 * Shape3D.pi * self._radius * (self._radius + self._height) 
        
    def __str__(self):
        print(super().__str__(), end = ", ")
        return "Height: " + str(self._height) + "\nArea: " + str(self.calc_surface_area())
                
    
sph = Sphere('Sphere', 5)
print('----------------------------------')
sph.calc_surface_area()
print(sph)
print('==================================')
cyl = Cylinder('Cylinder', 5, 10)
print('----------------------------------')
cyl.calc_surface_area()
print(cyl)
```

```python
#task 07





class PokemonBasic:
    def __init__(self, name = 'Default', hp = 0, weakness = 'None', type = 'Unknown'):
        self.name = name
        self.hit_point = hp
        self.weakness = weakness
        self.type = type
    def get_type(self):
        return 'Main type: ' + self.type
    def get_move(self):
        return 'Basic move: ' + 'Quick Attack'
    def __str__(self):
        return "Name: " + self.name + ", HP: " + str(self.hit_point) + ", Weakness: " + self.weakness
    
    
class PokemonExtra(PokemonBasic):
    def __init__(self, name, hp, weakness, type1, type2 = "Unknown", move = "Unknown"):
        super().__init__(name, hp, weakness, type1)
        self.type2 = type2
        self.move = move
        
    def get_type(self):
        if self.type2 != "Unknown":
            return super().get_type() + ", " + "Secondary type: " + self.type2
        else:
            return super().get_type()
                  
    def get_move(self):
        if self.move != "Unknown":
            return super().get_move() + "\n" + "Other move: " + self.move[0] + ", " + self.move[1]
        else:
            return super().get_move()
        

print('\n------------Basic Info:--------------')
pk = PokemonBasic()
print(pk)
print(pk.get_type())
print(pk.get_move())
print('\n------------Pokemon 1 Info:-------------')
charmander = PokemonExtra('Charmander', 39, 'Water', 'Fire')
print(charmander)
print(charmander.get_type())
print(charmander.get_move())
print('\n------------Pokemon 2 Info:-------------')
charizard = PokemonExtra('Charizard', 78, 'Water', 'Fire', 'Flying', ('Fire Spin', 'Fire Blaze'))
print(charizard)
print(charizard.get_type())
print(charizard.get_move())
```
