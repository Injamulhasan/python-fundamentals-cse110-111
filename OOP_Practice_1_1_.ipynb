Expected Output:

```python
class Team:
    def __init__(self, name):
        self.name = name
        self.players = []
   
    def addPlayer(self, *args):
        i = 0
        while (i < len(args)):
            p = args[i]
            
            if p.name == "":
                print('Invalid player Name')
                
            elif p.role == "":
                print('A player cannot be added without a role.')
                
            else:
                self.players.append(p)
                print(p.name + ' added in team')
                
            i = i + 1
        
    def showPlayers(self):
        bat = ""

        for i in range(0, len(self.players)):
            p = self.players[i]
            
            if p.role == 'Batsman':
                bat += 'Player Name: ' + p.name + ' Jersey No: ' + str(p.no) + '\n'
        
        if bat != "":
            print('Batsman: \n' + bat)
            
        bowl = ""
        
        for i in range(0, len(self.players)):
            p = self.players[i]
            
            if p.role == 'Bowler':
                bowl += 'Player Name: ' + p.name + ' No: ' + str(p.no) + '\n'
        
        if bowl != "":
            print('Bowler: \n' + bowl)
       
        
class Player:
    def __init__(self, name, role, no = ""):
        self.name = name
        
        if no == "":
            self.role = ""
            self.no = role
            
        else:
            self.role = role
            self.no = no
    
    def setRole(self, role):
        self.role = role

 

team = Team("Bangladesh")
p1 = Player("Mahmudullah", "Batsman",30)
print("=================================")
team.addPlayer(p1)
print("=================================")
team.showPlayers()
print("=================================")
p2 = Player("Mustafizur Rahman", "Bowler",90)
p3 = Player("Tamim Iqbal", 28)
print("=================================")
team.addPlayer(p2,p3)
print("=================================")
team.showPlayers()
print("=================================")
p3.setRole("Batsman")
print("=================================")
team.addPlayer(p3)
print("=================================")
team.showPlayers()
print("=================================")
p4 = Player("Mushfiqur Rahim", "Batsman",15)
p5 = Player("Taskin Ahmed", "Bowler",3)
print("=================================")
team.addPlayer(p4,p5)
print("=================================")
team.showPlayers()
```
